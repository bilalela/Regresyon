```python
df =pd.read_csv("community_health.csv")
df.head()
```

```python
df.corr(numeric_only=True)
```

```python
df.rename(columns={
        "age" : "yas",
        "gender" : "cinsiyet",
        "race/ethnicity" : "irk",
        "weight" : "kilo",
        "height" : "boy"

},inplace=True)
```

```python
df.head()
```

```python
df["kilo"] = round(df["kilo"] * 0.45)
```

```python
df["boy"] = round(df["boy"] * 2.54)
```

```python
df.head()
```

```python
df["cinsiyet"].replace({
           "female" : 0,
           "male" : 1             
},inplace= True)
```

```python
df["irk"].replace({
       "group A" : 1,
       "group B" : 2,
       "group C" : 3,
       "group D" : 4,
       "group D" : 5,
       "group E" : 6                  
},inplace=True)
```
