```python
def OP (power):
    if power < 320:
            return "*"
    elif power < 420:
         return "**"
    elif power < 500:
        return "***"
    else:
        return "#"
                                 
```

```python
df ["Star"] = df["Total"].apply(OP)
```

```python
df
```

```python
def attac(atak,Sp):
    if atak + Sp >240:
        return "Soword"
    else:
        return "Natural"
                                                                             
```

```python
df["Attackar"] = np.vectorize(attac)(df["Attack"],df["Sp. Atk"])
```
